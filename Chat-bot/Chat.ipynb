```python
import os
from mistralai import Mistral
import matplotlib.pyplot as plt 
```

## Mistral AI SAS is a French artificial intelligence (AI) company, headquartered in Paris. Founded in 2023, it has open-weight large language models (LLMs), with both open-source and proprietary AI models. 

```python
client = Mistral(api_key="API-kEY")
MODEL = "mistral-small-latest"
TEMPERATURE= 0.7#how creative will our chatbot answer the questions
MAX_TOKENS = 100 #Limits our token use
plt.barh(TEMPERATURE, MAX_TOKENS, color="pink", edgecolor="black")
plt.title("Creativity Over Tokens")
plt.show(TEMPERATURE, MAX_TOKENS)
SYSTEM_PROMPT = "You are a 90's programmer with ackward sense of humour revolving around computers"
messages = [{"role":"system","content":SYSTEM_PROMPT}]
```

```python
def chatbot(user_input):
    messages.append({"role": "user", "content": user_input})
    response= client.chat.complete(
        model = MODEL,
        message=messages,
        temperature= TEMPERATURE,
        max_tokens = MAX_TOKENS
    )
    reply = response.choices[0].message.content
    messages.append({"role": "assistant", "content":reply}) 
    return reply

```

 ## This is the API call

What happens behind the scenes:

Messages are serialized to JSON

Sent to Mistralâ€™s servers

AI processes conversation

Response is returned

```python
while True:
    user_input=input("You:")
    if user_input.strip().lower() in {"exit","quit"}:
        break
    
    answer = chatbot(user_input)
    print("Assistant:",answer)
    
```
